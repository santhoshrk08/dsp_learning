# Paper 1 — Week 1

**Title:** ECG Baseline Estimation and Denoising With Group 
Sparse Regularization

**Authors:** H. Shi, R. Liu, C. Chen, M. Shu and Y. Wang

**Year:** 2021

**IEEE Link:** https://ieeexplore.ieee.org/document/9344703

## 5-Bullet Summary

1. Problem: ECG signals get corrupted during recording by two 
   main interferences — random noise (unpredictable electrical 
   fluctuations) and baseline wander (slow drift caused by 
   patient breathing and body movement). Both distort the signal 
   and make diagnosis harder.

2. Dataset used: MIT-BIH Arrhythmia Database (PhysioNet) — 48 half-hour ECG recordings from real patients.

3. Method/Filter used: Group Sparse Regularization — a 
   mathematical optimization approach that separates baseline 
   wander from the true ECG signal by exploiting the sparse 
   (infrequent) structure of ECG beats. Different from 
   traditional filters like Butterworth.

4. Sampling frequency: 360 Hz — which satisfies Nyquist 
   for ECG signals (max content ~150 Hz, so need ≥ 300 Hz).

5. SNR improvement reported: Their method achieved higher 
   SNR improvement than traditional filters like moving 
   average and wavelet-based methods in comparison tests.
