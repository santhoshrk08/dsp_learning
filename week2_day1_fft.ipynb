```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import filtfilt,butter
```

```python
fs=1000
t=np.linspace(0,1,fs)
f1=5
f2=20
f3=50
y1=np.sin(2*np.pi*f1*t)
y2=np.sin(2*np.pi*f2*t)
y3=np.sin(2*np.pi*f3*t)
y_clean=y1+y2+y3
plt.figure(figsize=(10,4))
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,y_clean)
plt.grid(alpha=0.3)

plt.show()
```

```python
freqs = np.fft.fftfreq(len(y_clean), 1/fs)
fft_mag = np.abs(np.fft.fft(y_clean)) * 2 / len(y_clean)
plt.plot(freqs,fft_mag,color="red")
plt.xlim(-100,100)
plt.xlabel("Frequency (hz)")
plt.ylabel("Amplitude")
plt.grid(alpha=0.3)

plt.show()
```

```python
n=len(y_clean)
fft_values=np.fft.fft(y_clean)
fft_mag= np.abs(fft_values[:n//2])*1/n
freqs=np.fft.fftfreq(n,1/fs)[:n//2]
plt.plot(freqs,fft_mag,color="green")
plt.xlim(0,100)
plt.xlabel("Frequency (hz)")
plt.ylabel("Amplitude")
plt.grid(alpha=0.3)

plt.show()
```
